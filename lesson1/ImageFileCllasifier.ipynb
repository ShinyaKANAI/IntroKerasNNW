```python
def load_png(filepath):
    image = Image.open(filepath)
    image = ImageOps.grayscale(image)
    image = np.array(image)
    image = image / 255.
    image = image.astype(np.float32)
    return image
```

```python
img = load_png('kyomu.png')
plt.imshow(img, cmap='gray')
plt.show()
```
